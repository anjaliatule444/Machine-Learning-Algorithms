```python
## Import
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
iris=load_iris()
iris
```

```python
iris.DESCR
```

```python
## Creating Dependent and independent variable
x=iris.data
y=iris.target
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

```python
## X_train shape
X_train.shape
```

```python
## X_test shape
X_test.shape
```

```python
## Create and fit decision tree
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(X_train,Y_train)
```

```python
## Prediction
Y_pred=clf.predict(X_test)
```

```python
## Accuracy score
from sklearn.metrics import accuracy_score,r2_score,mean_absolute_error,mean_squared_error
print("R2 Score=",r2_score(Y_test,Y_pred))
print("Accuracy Score=",accuracy_score(Y_test,Y_pred))
print("Mean Absolute Error=",mean_absolute_error(Y_test,Y_pred))
print("Mean Squared Error=",mean_squared_error(Y_test,Y_pred))
print("Root Mean Squared Error=",np.sqrt(mean_squared_error(Y_test,Y_pred)))
```

```python
## Visualize tree
from sklearn.tree import plot_tree
from matplotlib.pylab import rcParams
rcParams['figure.figsize']=80,50

plot_tree(clf)
```

### Overfitting and Underfitting in decision trees

```python
import matplotlib.pyplot as plt
```

```python
## Function to depict Overfitting/underfitting
def analyzer(max_depth):
    ## Read Data
    data=pd.read_csv("H:\Full stack data science-Python pratice\Machine Learning\Machine-Learning-Algorithms\Datasets for Algorithms\Social_Network_Ads.csv")

    ## Get X as age and estimated salary--> For getting the values in numpy array(.values) is used
    X=data.iloc[:,2:4].values

    ## Get target
    Y=data.iloc[:,-1].values

    ##Create DT classifier and fit it
    clf=DecisionTreeClassifier(max_depth=max_depth)
    clf.fit(X,Y)

    ## Create a numpy array starting from 0th columns min-1 to 0th columns max+1 by step of 0.1
    a=np.arange(start=X[:,0].min()-1, stop=X[:,0].max()+1, step=0.1)

    ## Create a numpy array starting from 1st columns min-1 to 1st columns max+1 by step of 100
    b=np.arange(start=X[:,1].min()-1, stop=X[:,1].max()+1, step=100)

    ## Create numpy meshgrid
    XX,YY=np.meshgrid(a,b)

    ## Flatten the XX and YY and create array
    input_array=np.array([XX.ravel(),YY.ravel()]).T

    ## Do prediction on input_array which is our test data
    labels=clf.predict(input_array)

    rcParams['figure.figsize']=10,5
    plt.contourf(XX,YY,labels.reshape(XX.shape),alpha=0.5)
    plt.scatter(X[:,0],X[:,1],c=Y)
```

```python
## Overfitting
analyzer(max_depth=20)
```

```python
## Underfitting
analyzer(max_depth=1)
```
